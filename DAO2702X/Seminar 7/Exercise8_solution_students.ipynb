## Question 2

Consider the following dataset on Singapore private house properties.

````python
# Read the properties data into a dataframe and show some records of the properties data.

import pandas as pd

prop = pd.read_csv("properties2.csv")
prop
````

Details of the variables are given below:
- `project`: The name of the property project.
- `street`: The street name of the property.
- `type`: The type of the property: apartment or condominium.
- `district`: Location of the property, given as the district code of Singapore, details can be found [here](https://www.mingproperty.sg/singapore-district-code/).
- `segment`: Location of the property, given as the region segment of Singapore: Core Central Region (CCR), Rest of Central Region (RCR), or the Outside Central Region (OCR).
- `tenure`: The tenure type of the property: 99 years or freehold.
- `commence`: The commencing year of the property.
- `sale`: The sales type of the property: new sale or resale.
- `price`: The total price (in Singapore Dollars) of the property.
- `area`: The area of the property in square feet.
- `level`: The levels of the condos.
- `date`: The date of the transaction. 

1. What is the percentage of records that have the commencing year to be missing?

**Solution**:

````python
# Create a series of boolean values indicating whether the value is missing (True) or not missing (False).
prop["commence"].isnull()
````

````python
# Calculate the mean to get the percentage of missing values.
prop["commence"].isnull().mean()
````

**Or:** Use the Pandas series `isna()` method instead of `isnull()` method.

2. Create a new column called `price_psf` that is the price per square foot of these properties, i.e. the total price divided by the area.

**Solution**:

````python
# Divide the "price" column by the "area" column to get the price per area, and use it to create the "price_psf" column.
prop["price_psf"] = prop["price"] / prop["area"]
prop
````

3. How many records in the dataset are condos/apartments located in district 5 and their levels are higher than 12? You may use the starting level in the "level" column as the actual level.

**Solution**:

````python
# Use the first 2 characters (starting level) of the "level" column as the actual level.
level = prop["level"].str[:2].astype(int)
level
````

````python
# Create the 2 series representing the conditions needed.
cond1 = prop["district"] == 5
cond2 = level > 12

# Intersect the 2 conditions.
cond3 = cond1 & cond2

# Calculate the sum as it is the number of records that satisfy the 2 conditions.
cond3.sum()
````

4. What is the average price per square foot of all condos/apartments located in district 5 and their levels are higher than 12?

**Solution**:

````python
# Get the subset of the prop dataframe that satisfy the two conditions.
subset = prop.loc[cond3]
subset
````

````python
# Calculate the mean of the "price_psf" column.
subset["price_psf"].mean()
````

5. Consider all condos/apartments transactions that were done in 2020 or later, what is the median of the areas?

**Solution**:

````python
# Create a series 
# - with the last 2 characters of the string values in the "data" column
# - convert the 2 characters from string to integer
year = prop["date"].str[-2:].astype(int)

# Create a series of boolean values indicating whether the year is 2020 (20) or more.
cond = year >= 20
cond
````

````python
# Get the series of area values for records with year 20 or more.
# Then calculate the median of the areas.
prop.loc[cond, "area"].median()
````

6. Consider all transactions that were done in 2020 or later, what is the percentage of condos/apartments that are located in the RCR region?

**Solution**:

````python
# Get the series of segment values for records with year 2020 (20) or more.
# Get the number of records in each of the segments.
prop.loc[cond, 'segment'].value_counts()
````

````python
# Normalize the number of records in each of the segments to a proportion.
# They will sum to 1.
prop.loc[cond, 'segment'].value_counts(normalize=True)
````

7. Consider transaction records that are 1) resale condos with a 99-year tenure; 2) with prices between 2,000,000 and 2,200,000; and 3) sold in the year of 2021. Use one data visualization to show how the commencing years affect the price per square foot.

**Solution**:

````python
# Create series of boolean values for each of the 3 conditions.
cond1 = (prop['type'] == 'Condominium') & (prop['sale'] == 'Resale') & (prop['tenure'] == '99 years')
cond2 = (prop['price'] >= 2000000) & (prop['price'] <= 2200000)
cond3 = prop['date'].str[-2:] == "21"

# Create the subset of the prop dataframe that satisfy the 3 conditions.
subset = prop.loc[cond1 & cond2 & cond3]
subset
````

````python
# Plot "price_psf" against "commence" for the subset.

import matplotlib.pyplot as plt

plt.title("Price per sq ft vs Commencing year")
plt.xlabel("Commencing year", fontsize=12)
plt.ylabel("Price (S$/sqft)", fontsize=12)
plt.scatter(subset["commence"], subset["price_psf"], alpha=0.5)
plt.grid()
plt.show()
````

8. Consider transaction records that are 1) resale condos with a 99-year tenure; 2) with prices between 2,000,000 and 2,200,000; and 3) sold in the year of 2021. Use one data visualization that shows for each segment, how the commencing years affect the price per square foot.

**Solution**:

````python
# Plot "price_psf" against "commence" for each segment of the subset. 

import matplotlib.pyplot as plt

ocr = subset.loc[subset["segment"] == "OCR"]
rcr = subset.loc[subset["segment"] == "RCR"]
ccr = subset.loc[subset["segment"] == "CCR"]

plt.title("Price per sq ft vs Commencing year")
plt.xlabel("Commencing year", fontsize=12)
plt.ylabel("Price (S$/sqft)", fontsize=12)
plt.scatter(ocr["commence"], ocr["price_psf"], color="b", alpha=0.5, label="OCR")
plt.scatter(rcr["commence"], rcr["price_psf"], color="r", alpha=0.5, label="RCR")
plt.scatter(ccr["commence"], ccr["price_psf"], color="g", alpha=0.5, label="CCR")
plt.grid()
plt.legend(fontsize=12)
plt.show()
````
